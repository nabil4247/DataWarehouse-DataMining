```python
# Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

#bring in the six packs
df_train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Dataset-Week 11/Family Income and Expenditure.csv')
```

```python
#scatter plot Household Head Age/Total Income from Entrepreneurial Acitivites
var = 'Household Head Age'
data = pd.concat([df_train['Total Income from Entrepreneurial Acitivites'], df_train[var]], axis=1)
data.plot.scatter(x=var, y='Total Income from Entrepreneurial Acitivites', ylim=(0,800000));
```

```python
#box plot Household Head Age/Total Income from Entrepreneurial Acitivites
var = 'Household Head Age'
data = pd.concat([df_train['Total Income from Entrepreneurial Acitivites'], df_train[var]], axis=1)
f, ax = plt.subplots(figsize=(8, 6))
fig = sns.boxplot(x=var, y="Total Income from Entrepreneurial Acitivites", data=data)
fig.axis(ymin=0, ymax=800000);
```
